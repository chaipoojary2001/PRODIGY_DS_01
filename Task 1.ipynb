```python
import pandas as pd

# Load the main dataset
data = pd.read_csv('C:\\Users\\chaip\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_1538088.csv', skiprows=4)

# Load the metadata country file
metadata_country = pd.read_csv('C:\\Users\\chaip\\Downloads\\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_1538088.csv')
```

```python
year = '2023'
population_2023 = data[['Country Name', 'Country Code', year]].dropna()
```

```python
population_2023 = population_2023.merge(metadata_country[['Country Code', 'Region']], on='Country Code')
```

```python
population_by_region = population_2023.groupby('Region')[year].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 8))
population_by_region.plot(kind='bar', color='skyblue')
plt.title('Total Population by Region in 2023')
plt.xlabel('Region')
plt.ylabel('Total Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.hist(population_2023[year], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Population in 2023')
plt.xlabel('Total Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
